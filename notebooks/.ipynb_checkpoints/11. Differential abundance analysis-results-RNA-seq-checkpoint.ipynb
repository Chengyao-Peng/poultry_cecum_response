```python
%pip install pandas
%pip install openpyxl
import pandas as pd
import pickle
import os
HERE = os. getcwd()
DATA_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'data'))
RESULT_DIRECTORY = os.path.abspath(os.path.join(HERE, os.pardir, 'results'))
OUTPUT_PATH = os.path.join(RESULT_DIRECTORY, 'DeSeq2')
```

```python
outnames = ['Cecum_Digesta_03', 'Cecum_Mucosa_03', 'Colon_Digesta_03', 'Colon_Mucosa_03', 
            'Ileum_Digesta_03', 'Ileum_Mucosa_03', 'Swab_Marked_03', 'Swab_Sampled_03',
            'Cecum_Digesta_14', 'Cecum_Mucosa_14', 'Colon_Digesta_14', 'Colon_Mucosa_14', 
            'Ileum_Digesta_14', 'Ileum_Mucosa_14', 'Swab_Marked_14', 'Swab_Sampled_14',
            'Cecum_Digesta_21', 'Cecum_Mucosa_21', 'Colon_Digesta_21', 'Colon_Mucosa_21', 
            'Ileum_Digesta_21', 'Ileum_Mucosa_21', 'Swab_Marked_21', 'Swab_Sampled_21',
            'Cecum_Digesta_35', 'Cecum_Mucosa_35', 'Colon_Digesta_35', 'Colon_Mucosa_35', 
            'Ileum_Digesta_35', 'Ileum_Mucosa_35', 'Swab_Marked_35', 'Swab_Sampled_35']


list_CTR_AGP = ["CTR", "AGP"]
list_CTR_DC = ["CTR", "DC"]
list_CTR_DCandAGP = ["CTR", "DC-AGP"]
list_CTR_HBT = ["CTR", "HBT"]
list_CTR_DCandHBT = ["CTR", "DC-HBT"]

list_AGP_DC = ['AGP', 'DC']
list_AGP_DCandAGP = ['AGP', 'DC-AGP']
list_AGP_HBT = ['AGP', 'HBT']
list_AGP_DCandHBT = ['AGP', 'DC-HBT']


list_DC_DCandAGP = ["DC", "DC-AGP"]
list_DC_HBT = ["DC", "HBT"]
list_DC_DCandHBT = ["DC", "DC-HBT"]

list_DCandAGP_HBT = ["DC-AGP", "HBT"]
list_DCandAGP_DCandHBT = ["DC-AGP", "DC-HBT"]

list_HBT_DCandHBT = ["HBT", "DC-HBT"]

full_list=[list_CTR_AGP,list_CTR_DC,list_CTR_DCandAGP,list_CTR_HBT,list_CTR_DCandHBT,
           list_AGP_DC,list_AGP_DCandAGP,list_AGP_HBT,list_AGP_DCandHBT,
           list_DC_DCandAGP, list_DC_HBT, list_DC_DCandHBT,
           list_DCandAGP_HBT,list_DCandAGP_DCandHBT,list_HBT_DCandHBT]
```

# 1. Day 03

```python
for outname in outnames:
    for each_pair in full_list:
        if not outname+'_'+each_pair[1]+'_'+each_pair[0] in ['Colon_Digesta_03_DC_CTR']:
            try:
                shrunk_stat_result=pd.read_excel(os.path.join(OUTPUT_PATH, outname+'_'+each_pair[0]+'_'+each_pair[1]+"_shrunk_stat_results.xlsx"), index_col=0)
            except FileNotFoundError:
                shrunk_stat_result=pd.read_excel(os.path.join(OUTPUT_PATH, outname+'_'+each_pair[1]+'_'+each_pair[0]+"_shrunk_stat_results.xlsx"), index_col=0)
            
            sig_result = shrunk_stat_result[shrunk_stat_result['padj']<0.01]
            
            with open(os.path.abspath(os.path.join(HERE, os.pardir, 'results', 'DeSeq2', 'DAA_sig.txt')), 'a') as file:
                file.write(f"{outname}\n")
                file.write(each_pair[0] + ' VS ' + each_pair[1] + "\n")
                
                print(outname + "\t" + each_pair[0] + '_VS_' + each_pair[1] + "\t" + str(sig_result.shape[0]))
            
                for index, row in sig_result.iterrows():
                    value = row['padj']
                    file.write(f"{index}\t{value}\n")
                file.write("\n")

        
```

```python
sig_numbers = pd.read_csv(os.path.abspath(os.path.join(HERE, os.pardir, 'results', 'DeSeq2', 'DeSeq2_sig_numbers.txt')), sep="\t", names=['Name', 'Comparison','Sig number'])
```

```python
def remove_three_chars(s):
    if len(s) > 3:
        return s[:-3]  # Remove the last three characters
    else:
        return '' 
```

```python
sig_numbers[['Gut_section', 'Measure_type', 'Age']] = sig_numbers.Name.str.split('_', expand=True)
sig_numbers['Name'] = sig_numbers['Name'].apply(remove_three_chars)
sig_numbers
```
